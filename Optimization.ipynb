# What is optimization?
Mathematical optimization (aka mathematical programming) is the selection of a *best* element from some set of available alternatives.

An optimization problem consists of minimizing (maximizing) a real function by choosing input values from within an allowed set. 

More generally, optimization includes finding *best available* values of some objective function, given a defined domain (or input), where objective functions and domains can be very diverse. 

A lot of problems in AI are based on optimization! All machine learning and deep learning for example have three steps:
- 1 define a model
- 2 define a function loss (in general called objective function, because can be also a gain function)
- 3 minimize the function loss = traininig = fit = find the best parameters of the model.

Usually we don't have a closed formula to find the solution, so are needed iterative methods like the gradient descendent, that's why we talk about learning. Not only that reason: The difference between optimization and machine learning is this: while optimization algorithms can minimize the loss on a training set, machine learning model allows us to do predictions on unseen examples.

# Classification of optimization problems:
(the term "programming" referse to "planning", it doesn't refer to "writing with a programming language". You can also replace the word programming with "optimization problems").
- Convex programming: the objective function is convex (minimization) or concave (maximization) and the constraint set is convex.
   - Linear programming (LP): the objective function f is linear and the constraints are specified using only linear equalities and inequalities.
      - Integer programming: linear programs in which some or all variables are constrained to take on integer values (this is in general much more difficult than regular linear programming).
- Nonlinear programming (NLP): the objective function and/or the constraints can be nonlinear. (can be convex or not).
- Quadratic programming: the objective function can have quadratic terms, while the constraits are linear. (can be convex or not)

- Combinatorial optimization: problems where the set of feasible solutions is discrete or can be reduced to a discrete one.
- Stochastic programming: some of the constraints or parameters depend on random variables. (for example a descent method where some alpha or rho are chosen using a random variable in the process).
- Robust programming: tries to capture uncertainty in the data, finding solutions that are valid under all / a wide set of possible realizations of the uncertainties.
- Constraint programming: a programming paradigm where relations between variables are stated in the form of constraints.
. . . and many more . . . 

# How problems are solved:
To solve problems, researchers may use algorithms that terminate in a finite number of steps, or iterative methods that converge to a solution (on some specified class of problems), or heuristics that may provide approximate solutions to some problems (although their iterates need not converge).

- Optimization algorithms
   - Simplex algorithm (for linear programming, or an extension of it for quadratic programming).
   - Combinatorial algorithms
   - Quantum optimization algorithms
   
- Iterative methods: needed to solve non-linear problems. In particular for as regard optimization we are interested in "DESCENT METHODS" which are based on this idea:
  We start from a starting point $\vec{x}_o$ and we want at each iteration find another x in which the function has a lower     value. So at each step we want $f(\vec{x}_{k+1}) < f(\vec{x}_k)$. In general the descent methods do this at each iteration:
  $$ \vec{x}_{k+1} = \vec{x}_k + \alpha_k \vec{\rho}_k $$
  where $\alpha$ is the "step length" or "learning rate". Instead $\rho$ is the "descent direction" vector, which has the same dimension of $\vec{x}$. We'll discuss the main methods later on, for the moment let's classify them. They differ on the way $\alpha$ and $\rho$ are computed. 
  
    about $\alpha$:

  - constant
  - EXACT LINE SEARCH methods: at each step is sought the step length which minimizes the function along the given direction. They have an higher cost of course.
  - INEXACT LINE SEARCH (= "BRACKTRACKING"): at each iteration step there's an additional iterative procedure which stops when the value of alpha satisfies two properties: 1)  $f(\vec{x}_{k+1}) < f(\vec{x}_k)$ 2) a convergence test.
  
  about $\rho$:
      
  - FIRST ORDER METHODS: gradient methods: evaluate gradients, or approximate gradients in some way (or even subgradients)
    usually each step is faster, but it could converge in much more time (Q-LINEAR SPEED):
  
      - Gradient descent (alternatively, "steepest descent" or "steepest ascent"): A (slow) method of historical and theoretical interest, which has had renewed interest for finding approximate solutions of enormous problems.  
      - Quasi-Newton methods: Iterative methods for medium-large problems (e.g. N<1000).
      - Gauss-Newton and Levenberg-Marquardt, they are modifications of the Newton's method in order to do not compute the Hessian matrix! But they can be used only to minimize a sum of squared function values (indeed are mostly used for non linear square problems).
      - ... 
  - SECOND ORDER METHODS: Newton like methods: evaluate Hessians (or approximate Hessians, using finite differences)
    Usually each step is more costly, but it has an higher convergence speed (Q-QUADRATIC SPEED):

    - Newton's method 
    - ...

- A heuristic is any algorithm which is not guaranteed (mathematically) to find the solution, but which is nevertheless useful in certain practical situations. List of some well-known heuristics:
   - hill climbing
   - genetic algorithms
   - ant colony
   - articial bee colony
   - PSO
   - simulated annealing

# Derivative rules multivariate case:
In the multicase we have different kind of derivatives, let's discuss them considering a function $f : \mathbb{R}^n \to \mathbb{R}^m$
- PARTIAL DERIVATIVE: Is a derivative along the direction of variation of only one variable!! The others are considered constants! It's usually denoted like this: $$\frac{\partial {f(\vec{x})}}{\partial x_3}$$
  How can be computed? By using the univariate rules, considering the other variables as constants.
  
  
- DIRECTIONAL DERIVATIVE: Is a derivative along a direction specified by a vector v. In this direction many variables can have variations, not only one! It's usually denoted like this: $$\frac{\partial {f(\vec{x})}}{\partial \vec{v}} \quad or \quad f'_v(\vec{x}) \quad   or  \quad \nabla _v f(\vec{x})$$
  How can be computed? By exploiting the gradient property, which says that the derivative along a direction can be obtained by doing the dot product of the gradient and the vector describing the direction: $${\displaystyle \nabla _{\mathbf {v} }{f}(\mathbf {x} )=\nabla f(\mathbf {x} )\cdot \mathbf {v} } $$
  In order to make the value of the derivative indipendent from the length of the vector, is in general used to use normalied vector (simply divide it for its module).


# CHAIN RULE
This is really important for deep learning, so let's discussi it more deeply:

Let x be a real number, and let f and g both be functions mapping from a real number to a real number. 
Suppose that y = g(x) and z = f(g(x)) = f(y). Then the chain rule states that:
$$\frac{dz}{dx} = \frac{dz}{dy} \frac{dy}{dx} $$

We can generalize this beyond the scalar case. Suppose that $x \in \mathbb{R}^m$ , $ y \in \mathbb{R}^n$, $g:\,\,  \mathbb{R}^m \to \mathbb{R}^n$ , and $f:\,\, \mathbb{R}^n \to \mathbb{R} $. If y = g(x) and z = f(y), then:

$$ \frac{\delta z}{\delta x_i} = \sum_j \frac{\delta z}{\delta y_j} \frac{\delta y_j}{\delta x_i} $$

In vector notation, this may be equivalently written as:

$$ \nabla_\vec{x} z =  \big(\frac{\delta \vec{y}}{\delta \vec{x}}\big)^T \nabla_\vec{y} z $$
where $ \big(\frac{\delta \vec{y}}{\delta \vec{x}}\big) $ is the n x m Jacobian matrix of g.

# OPTIMIZATION OF A MULTIVARIATE FUNCTION $f(x_1,x_2,..x_n) = value$

to find a stationary point / critical point (= maximum, minimum or saddle point) $\vec{x}^∗ =[x^∗_1,x^∗_2,...,x^∗_n]^T$ :

$ \frac{\partial f}{\partial x}\Big|_{x=x_i} = \vec{0} \quad  \forall i=1..n$  which can also be expressed as:

${\displaystyle \nabla f(\vec{x})={\begin{bmatrix}{\frac {\partial f}{\partial x_{1}}}(\vec{x}) &,& \cdots &,& {\frac {\partial f}{\partial x_{n}}}(\vec{x})\end{bmatrix}}.} = GRADIENT\, of\, f = \text{particular case of}\,JACOBIAN \,\text{of f }= \vec{0}.$

Note that the gradient is a row vector!!! 

but it's more difficult to estabilish if it is a minimum or a maximum: we need to consider the second derivative for all possible pairs of i and j. So we need to talk about the Hessian matrix:
- If $H(\vec{x}^∗)$ is POSITIVE DEFINITE, then $\vec{x}^∗$ is a MINIMUM stationary point.
- If $H(\vec{x}^∗)$ is NEGATIVE DEFINITE, then $\vec{x}^∗$ is a MAXIMUM stationary point.
- If $H(\vec{x}^∗)$ is INDEFINITE, then $\vec{x}^∗$ is a SADDLE stationary point (can have a minimum value in one direction, and a maximum value in another)

where ${\displaystyle H(\vec{x}^*){\text{ positive definite}}\quad \iff \quad \vec{z}^{\textsf {T}}H(\vec{x}^*)\vec{z}>0{\text{ for all }}\vec{z}\in \mathbb {R} ^{n}\setminus \mathbf {0} } \qquad \text{where}:$

- $\vec{x}^*$ is the critical point.
- $H(\vec{x}^*)$ is the Hessian matrix in which the correspective variables have been substituted with the values of the critical point considered.
- $\vec{z}$ is a generic vector $\in \mathbb {R} ^{n}\, ,\,\neq \vec{0}$

(obviously positive semi-definite means that the result can be >=0. And it's trivial the meaning of negative definite and semi-definite)
# Hessian matrix:


$$
\begin{bmatrix} 
\frac{\partial^2f}{\partial x_1\partial x_1}  & \frac{\partial ^2f}{\partial x_1 \partial x_2} & \cdots & \frac{\partial ^2f}{\partial x_1 \partial x_n} \\ 
\frac{\partial ^2f}{\partial x_2\partial x_1}  &\frac{\partial ^2f}{\partial x_2\partial x_2} & \cdots & \frac{\partial ^2f}{\partial x_2 \partial x_n} \\ 
\vdots & \vdots & \ddots & \vdots \\  
\frac{\partial ^2f}{\partial x_n\partial x_1} & \frac{\partial ^2f}{\partial x_n\partial x_2} & \cdots & \frac{\partial ^2f}{\partial x_n\partial x_n}
\end{bmatrix}
=
\begin{bmatrix} 
11  & 12 & \cdots & 1n\\ 
21 & 22 & \cdots & 2n \\ 
\vdots & \vdots & \ddots & \vdots \\ 
n1 & n2 & \cdots & nn
\end{bmatrix}
$$

(Easy to remember: the variables involved are exactly those which indicates the position ij in the matrix. Anyway H is symmetric so it's not so important.)

Note:

-  ${\frac  {\partial f }{\partial x_{i} \partial x_{j}}}= {\frac  {\partial }{\partial x_{i} }}\left({\frac  {\partial f}{\partial x_{j}}}\right)={\frac  {\partial }{\partial x_{j}}}\left({\frac  {\partial f}{\partial x_{i}}}\right).$

- H is always symmetric.

- The Hessian matrix of a function f is the transpose of the Jacobian matrix of the gradient of the function: $H(f(x)) = J(∇f(x))^T.$

- the trace of H is called "laplacian operator".

# OPTIMIZATION OF VECTOR-VALUED FUNCTIONS:
A vector function is a function $f : \mathbb{R}^n \to \mathbb{R}^m$ where m>0.
It can be expressed in these forms:

${\mathbf  {f}}(\mathbf {x})=( f_1(\mathbf {x}),f_2(\mathbf {x}),...,f_m(\mathbf {x}))\quad\quad where \quad x \in \mathbb{R}^n$

For this kind of functions we must generalize the concepts:

Hessian matrix --> Hessian tensor

Gradient --> Jacobian

more details:

# Hessian tensor:

The collection of second partial derivatives of a vector-valued function is not a n×n matrix, but rather a third order tensor. This can be thought of as an array of m Hessian matrices, one for each component of f:

$${\displaystyle \mathbf {H} (\mathbf {f} )=(\mathbf {H} (f_{1}),\mathbf {H} (f_{2}),\dots ,\mathbf {H} (f_{m}))}$$


# Jacobian
It the collector of all the first order derivatives of the function. In case of functions $f : \mathbb{R}^n \to \mathbb{R}$ the Jacobian is also called gradient. In case of vector valued functions $f : \mathbb{R}^n \to \mathbb{R}^m$ it has this general form:


$$
\begin{bmatrix}
\frac{\partial f_1}{\partial x_1} & \cdots &
\frac{\partial f_1}{\partial x_n} \\
\vdots & \ddots & \vdots \\
\frac{\partial f_m}{\partial x_1} & \cdots &
\frac{\partial f_m}{\partial x_n}
\end{bmatrix}
=
\begin{bmatrix}
\nabla f_1(\vec{x}) \\
\vdots  \\
\nabla f_m(\vec{x})
\end{bmatrix}
$$



- which can be seen as a matrix which has for each row the gradient of the function $f_i$
- note that in case of a function $f : \mathbb{R} \to \mathbb{R}^m$ the Jacobian is a column vector. Instead when the function is $f : \mathbb{R}^n \to \mathbb{R}$ we have the gradient, which is a row vector.

\end{bmatrix} $$


# Taylor series for univariate functions
A Taylor series is a representation of an infinitely differentiable function as an infinite sum of terms that are calculated from the values of the function's derivatives at a point c:
$$ f(x) = f(c) + \frac{f'(c)}{1!} (x-c)+ \frac{f''(c)}{2!}(x-c)^2 + ... = \sum_{k=0}^\infty \frac{f^n(c)}{n!}(x-c)^n$$

Note that with infinite terms we have an equivalence. If instead we use only a certain number of terms we have an approximation. The closer to the point c, the better the approximation.



# example taylor approximation: $ f(x) = 3x^5 -5x^3 $ at c= 0.5

$$ T(x)= -\frac{17}{32} + (-2\frac{13}{16})(x-0.5) + \frac{1}{2!}(-\frac{7}{2})(x-0.5)^2 + ...$$



# Taylor series for multivariate functions
As in the previous case, but the first derivative will be the gradient, while the second derivative the Hessian matrix. Ussually we can stop to the second term for optimization problems ("quadratic approximation"). But there's another difference: in the univariate case (x-c)^2 is a scalar, instead in the multivariate case we have $(\vec{x}-c)^2$ which isn't a possibe operation, and we also have to think how this can be multiplied to the Hessian which is a matrix nxn. Solution:
$$ f(\vec{x}) = f(c) + \nabla(f(c))(\vec{x}-c) + \frac{1}{2!}(\vec{x}-c)^TH(\vec{x}-c)+...$$

Let's explode the formula in the case of  $f : \mathbb{R}^2 \to \mathbb{R}$:

$$ f(\vec{x}) = f(c) + (\frac {\partial f}{\partial x_{1}}(c)\,,\, \frac {\partial f}{\partial x_{1}}(c)) \begin{pmatrix}x_1-c \\x_2-c\end{pmatrix} + \frac{1}{2!} (x_1 -c\,,\, x_2-c) \begin{bmatrix} 
\frac{\partial ^2f}{\partial x_1\partial x_1}  & \frac{\partial ^2f}{\partial x_1\partial x_2}  \\ 
\frac{\partial ^2f}{\partial x_2 \partial x_1}  &\frac{\partial ^2f}{\partial x_2\partial x_2}
\end{bmatrix} \begin{pmatrix}x_1-c \\x_2-c\end{pmatrix}  = $$

$$ f(\vec{x}) = f(c) + \frac {\partial f}{\partial x_{1}}(c) (x_1-c) + \frac {\partial f}{\partial x_{2}}(c) (x_2-c)  + \frac{1}{2!}\frac {\partial^2 f}{\partial x_1^2}(c)(x_1-c)^2 + \frac{2}{2!}\frac {\partial^2 f}{\partial x_{1}\partial{x_2}}(c)(x_1-c)(x_2-c) + \frac{1}{2!}\frac {\partial^2 f}{\partial x_2^2}(c)(x_1-c)(x_2-c) $$ 



# More definitions:
# CONVEX FUNTCION
A real-valued function defined on an n-dimensional interval is called convex (concave) if the line segment between any two points on the graph of the function lies above (below) or on the graph. 

In formula: a function $f: \mathbf{R}^n \to \mathbf{R} \quad \text{is convex if}\quad:$
$$ \forall \vec{x}_a,\vec{x}_b \in \mathbf{R}^n, \forall t \in [0,1]: \quad\,\,\, f(t\vec{x}_a+ (1-t)\vec{x}_b)\,\, \leq \,\,tf(\vec{x}_a) + (1-t)f(\vec{x}_b) $$

(Generalisation of this is trivial: instead of a segment we talk about a hyperplane).
(You can check if f is convex in a certain region, but the region in which you check must be a convex set)

If f is a twice-differentiable function we can test convexity in this way:
- For a function of a single variable, if its second derivative is always > = 0.
- In general: if the Hessian matrix is semipositive definite.

(indeed if you think about it remember that the second derivative describes the behaviour of the first derivative: if the second derivative is >=0 it means that the first derivative is always getting bigger; indeed in the example it starts from a high negative value at the left extreme, and becomes an high positive value on the right extreme).



Why do we care about this? Because are easier to do optimization, so we want to have loss functions which have this property. 

Why is it easier to do optimization with convex functions? Because if you find a minimum you're sure that it is a global minimum.

Of course we are interested also in finding the maximum of concave functions, but it's easy: a function is concave if its negation is convex.

In a convex problem, if there is a local minimum that is interior (not on the edge of the set of feasible elements), it is also the global minimum

# Inner/dot/scalar product 
It is an algebraic operation that takes two equal-length sequences of numbers (often coordinate vectors) and returns a single number. 

It can be computed in 2 ways:

# Line equations:
# line equation given  two points A,B.
The line equation is given by considering a generic point P(x,y) along the line,  so this property holds:
$$ \vec{AP} \lambda = \vec{AB} $$
so for example in the two dimensional case:
$$ (x - x_a, y - y_a ) \lambda = (x_b - x_a, y_b - y_a) $$ 
so:
$$
\begin{cases}
(x - x_a) \lambda = (x_b - x_a) \\
(y - y_a)  \lambda = (y_b - y_a)
\end{cases}
\quad \to \quad
\begin{cases}
 \lambda = \frac{(x_b - x_a)}{(x - x_a)} \\
 \lambda = \frac{(y_b - y_a)}{(y - y_a)}
\end{cases}
\quad \to \quad
\frac{(x_b - x_a)}{(x - x_a)} = \frac{(y_b - y_a)}{(y - y_a)}
\quad \to \quad
\frac{(x - x_a)}{(x_b - x_a)} = \frac{(y - y_a)}{(y_b - y_a)}
$$

By the way it's even easier to compute the parametric equation:
$$
\begin{cases}
(x - x_a) \lambda = (x_b - x_a) \\
(y - y_a)  \lambda = (y_b - y_a)
\end{cases}
\quad \to \quad
\begin{cases}
x   = x_a + t (x_b - x_a) \\
y   = y_a + t (y_b - y_a)
\end{cases}
$$

# line equation given a point $A(x_a,x_b)$ and a vector normal to the line $\vec{n} (a,b)$
considering a generic point P(x,y) along the line then the vector $\vec{AP}$ must be orthogonal to $\vec{n}$, so the dot product must be equal to zero: $$\vec{AP}\cdot\vec{n} = (x - x_a, y-y_a) \cdot (a,b) = a(x-x_a) + b(y-y_a) = ax +by - ax_a -by_a = ax +by + c= 0$$

Note that when you have the line equation in this form (the implicit form) it's trivial to know the vector normal to it.

# Plane equation given 3 points A,B,C:
We exploit the fact that 3 vectors belonging to the same plane must be linearly dependent (because a plane lives in $\mathbb{R}^2$), so their mist product is equal to zero: $$\vec{AB}\cdot\vec{AC} x \vec{BC} = 0$$

# Plane equation given a point A and a normal vector $\vec{n} = (a,b,c)$ :

Let's consider a generic point P of the plane. The vector $\vec{AP} = (x-x_a, y - y_a, z - z_a) $ must be orthogonal to $\vec{n}$ so:
$$ \vec{AP}  \cdot \vec{n} = (x-x_a, y - y_a, z - z_a) \cdot (a,b,c) = a(x-x_a)+ b(y - y_a) + c(z - z_a) = ax+ by+ cz + d = 0 $$

USEFUL PROPERTIES:
A plane of equation $ π:ax+by+cz+d=0 $  has direction normal to the vector $\vec{n}=(a,b,c)$ and the vector lying in the plane are all the vectors which satisfy this equation: $ ax+by+cz=0$.
